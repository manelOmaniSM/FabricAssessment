```python
Wksp_users = spark.sql("SELECT * FROM Workspace_users")
Wksp_users= Wksp_users.select('UserPrincipalName','CompanyName').dropDuplicates()
```

```python
CompanyMap= spark.sql("SELECT * FROM Company_Mapping").select('Company','OpCo').dropDuplicates()
```

```python
RLS_users=Wksp_users.join(CompanyMap,Wksp_users.CompanyName==CompanyMap.Company , "left_outer")\
                    .select("UserPrincipalName","OpCo")
```

```python
RLS_users.write.format('delta')\
.mode("overwrite")\
.option("overwriteSchema", "True")\
.save("Files/RLS")
```

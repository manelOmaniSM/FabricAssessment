```python
from pyspark.sql.functions import *

Wksp_users = spark.sql("SELECT * FROM Workspace_users")
Wksp_users=Wksp_users.filter(col("AccessRights")=="Admin")
```

```python
Wksp_users= Wksp_users.groupBy('WorkspaceID','CompanyName').count()
```

```python
from pyspark.sql.window import Window
windowPartition = Window.partitionBy("WorkspaceID").orderBy(desc("count"))

Wksp_users = Wksp_users.withColumn("rank",rank().over(windowPartition))\
                       .filter(col("rank")==1)\
                       .select("WorkspaceID","CompanyName")
```

```python
Wksp_users.write.format('delta')\
.mode("overwrite")\
.option("overwriteSchema", "True")\
.save("Files/Workspace_Company_Map")
```
